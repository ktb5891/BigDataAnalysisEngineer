```python
# alcohol(음주수치)이 상위 10번째 값으로 상위 1~10위의 범위의 값을 변경한 후
# speeding(속도)가 7이상의 alcohol(음주수치)데이터들의 평균 산출
# 최종 결과만 print 출력
```

```python
# data 불러오기
import seaborn as sns
import pandas as pd
sns.get_dataset_names()
df = sns.load_dataset('car_crashes')
print(df.head())
```

```python
data = df.sort_values(by = df.columns[2], ascending = False) # alcohol을 기준으로 내림차순 정렬
# data.head()
ten = data.iloc[9,2] # alcohol의 10번째 값
data.iloc[:10,2] = ten # 1~10위까지의 alcohol 값 변경
# data.head(11)
result = data[(data['speeding'] >= 7)] # speeding이 7 이상인 값
print('최종 결과물 : ', result['alcohol'].mean()) # 평균값 도출
```

```python
# 데이터 첫번째 행부터 70%까지의 데이터를 추출한 후
# distance(거리)의 결측값을 distance(거리)의 중앙값으로 대체하고
# 결측값 전처리 전과 후의 표준편차를 비교
# 전/후 결과값 print 출력
```

```python
# 데이터 불러오기
import seaborn as sns
import pandas as pd
import numpy as np

df = sns.load_dataset('planets')
df.head()
```

```python
length = int(len(df)*0.7) 
df = df[:length] # 70% 데이터 추출
```

```python
df2 = df.copy() 
df2['distance'] = df2['distance'].fillna(df2['distance'].median()) # 결측치를 중앙값으로 대체
df2
```

```python
print('전처리 전 distance 표준편차: ',np.std(df['distance']))
print('전처리 후 distance 표준편차: ',np.std(df2['distance']))
```

```python
# 통계 기법(IQR) 방식의 이상치 제거
```

```python
# orbital_period열의 이상치를 IQR 기법으로 제거하여
# 이상치들의 합을 계산
# 최종 결과만 print 출력
```

```python
# 데이터 불러오기

import seaborn as sns
sns.get_dataset_names()
df = sns.load_dataset('planets')
df.head()
```

```python
q25 = df['orbital_period'].quantile(0.25) # orbital_period의 1사분위 값
q75 = df['orbital_period'].quantile(0.75) # orbital_period의 3사분위 값
iqr = q75-q25 # iqr의 범위
print(q25, q75, iqr)
```

```python
min = q25-iqr*1.5 # 이상치를 제거한 최소 값
max = q75+iqr*1.5 # 이상치를 제거한 최대 값
print(min,max)
```

```python
df_outlier = df[(df['orbital_period'] <= min) | (df['orbital_period'] >= max)] # 이상치를 제거한 df
print(df_outlier['orbital_period'].sum())
```

```python
# 작업유형 2 수집 및 전처리
```

```python
# 데이터 불러오기

import seaborn as sns
df = sns.load_dataset('penguins')
df.head()
```

```python
# 데이터 결측치 보정

missing = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
for i in missing:
    df[i] = df[i].fillna(df[i].median())
df['sex'] = df['sex'].fillna('Male')
df.head()
```

```python
# 라벨 인코딩

label = ['species','island','sex']
from sklearn.preprocessing import LabelEncoder
df[label] = df[label].apply(LabelEncoder().fit_transform)
df
```

```python
# 데이터 타입, 더미 변환

import pandas as pd
category = ['island','sex']
for i in category:
    df[i] = df[i].astype('category')
df = pd.get_dummies(df)
df.head()
```

```python
# 파생변수 생성

df['body_mass_g_qcut'] = pd.qcut(df['body_mass_g'], 5, labels = False)
df.head()
```

```python
# 정규화 또는 스케일 작업

from sklearn.preprocessing import MinMaxScaler
scaler = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
min = MinMaxScaler()
min.fit(df[scaler])
df[scaler] = min.transform(df[scaler])
df.head()
```

```python
# 데이터 분리

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df['species'], test_size = 0.2, stratify = df['species'], random_state = 1)
print(X_train, X_test, y_train, y_test)
```

```python
# 모형 학습(1)

from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(max_depth = 6, n_estimators = 100)
model1.fit(X_train, y_train)
pred1 = model1.predict(X_test)
pred1
```

```python
# 모형 학습(2)

from sklearn.ensemble import AdaBoostClassifier
model2 = AdaBoostClassifier()
model2.fit(X_train, y_train)
pred2 = model2.predict(X_test)
pred2
```

```python
# 앙상블

from sklearn.ensemble import VotingClassifier
clf = VotingClassifier(estimators = [('rf',model1), ('ad',model2)], voting = 'hard')
clf.fit(X_train, y_train)
pred3 = clf.predict(X_test)
pred3
```

```python
# 모델 평가

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
print('랜덤 포레스트 정확도 : ', accuracy_score(y_test, pred1))
print('아이다 부스트 정확도 : ', accuracy_score(y_test, pred2))
print('앙상블(랜포+아이다) 보팅 정확도 : ', accuracy_score(y_test, pred3))
```

```python
# 하이퍼 파라미터 튜닝

from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators':[50,100], 'max_depth':[4,6]}
model4 = RandomForestClassifier()
clf = GridSearchCV(estimator = model4, param_grid = parameters, cv = 3)
clf.fit(X_train,y_train)
print('최적의 파라미터 : ', clf.best_params_)
```
